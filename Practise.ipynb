```python
def null_replace():
  # write your code here
  data['network'] = data['network'].fillna(data['network'].value_counts().index[0])
  data['network_type'] = data['network_type'].fillna(data['network_type'].value_counts().index[0])
  no_network = data['network'].value_counts()
  no_network_type = data['network_type'].value_counts()
  return no_network,no_network_type
null_replace()
```

#4. Split the month column to year and month_no and display how many entries are there in 2014 - 1 Marks

```python
def month():
  # write your code here
  data[['year','month_no']]=data.month.str.split("-",expand=True)
  entries = data.loc[data['year']=='2014']
  count = entries.shape[0]

  return count
month()
```

```python
def avg_2015():
  #write your code here
  #write your code here
  avg= data.loc[(data['year'] == '2015'),['duration']]
  avg = avg.mean()
  return round(avg,2)
avg_2015()
```

#6. What is the sum of durations, for calls only, to each network - 2 Marks

```python
def sum_network():
  #write your code here
  Meteor = data[data['network'] == 'Meteor']['duration'].sum()
  Tesco =  data[data['network'] == 'Tesco']['duration'].sum()
  Vodafone =  data[data['network'] == 'Vodafone']['duration'].sum()
  return round(Meteor,1),round(Tesco,1),round(Vodafone,1)
sum_network()
```

```python
def december_landline():
  #write your code here
  data1 = data.loc[(data["month_no"]=="12")&(data["network"]=="landline")]
  val = data1["duration"].count()
  return val
december_landline()
```
