### 2.2 Filling with central tendencies

```python
mean_age = dummy_df['Age'].mean()
mean_height = dummy_df['Height(cm)'].mean()
```

```python
map_dict = {'Age': mean_age, 'Height(cm)': mean_height}
map_dict
```

```python
dummy_df.fillna(value=map_dict)
```
